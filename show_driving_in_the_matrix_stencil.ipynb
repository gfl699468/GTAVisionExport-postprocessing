## shows stencil calculated by Driving in the Matrix stencil calculation

```python
import visualization
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
visualization.multi_page = False
visualization.ini_file = 'gta-postprocessing.ini'
```

```python
name = '2018-01-25--16-58-50--412'
TIFF_DIR = os.path.join('Y:\\', 'GTA-sample-data')
annotations_dir = os.path.join('Y:\\', 'GTA-sample-data-annotated')

conn = visualization.get_connection()
cur = conn.cursor()
cur.execute("""SELECT snapshot_id FROM snapshots WHERE imagepath = '{}'""".format(name))
# print(size)
snapshot_id = list(cur)[0]['snapshot_id']
pp_file = np.load(os.path.join(annotations_dir, '{}.npz'.format(snapshot_id)))
postprocessing_stencil = pp_file['arr_0']
im = Image.open(os.path.join(TIFF_DIR, name + '.tiff'))
```

```python
depth = visualization.load_depth(name)
stencil_ids = visualization.load_stencil_ids(name)

fig = plt.figure(figsize=(15,15))
plt.axis('off')
plt.imshow(im)

fig = plt.figure(figsize=(15,15))
plt.axis('off')
plt.imshow(depth, cmap='gray')

fig = plt.figure(figsize=(15,15))
plt.axis('off')
plt.imshow(visualization.ids_to_greyscale(stencil_ids), cmap='gray')

fig = plt.figure(figsize=(15,15))
plt.axis('off')
plt.imshow(postprocessing_stencil, cmap='gray')

plt.show()
```

```python
fig = plt.figure(figsize=(15, 15))
plt.axis('off')
plt.tick_params(axis='both', left='off', top='off', right='off', bottom='off', labelleft='off', labeltop='off',
                    labelright='off', labelbottom='off')
plt.imshow(im)
plt.imshow(stencil_ids == 2, cmap='bwr', alpha=.2)
plt.show()

fig = plt.figure(figsize=(15, 15))
plt.axis('off')
plt.tick_params(axis='both', left='off', top='off', right='off', bottom='off', labelleft='off', labeltop='off',
                    labelright='off', labelbottom='off')
plt.imshow(im)
plt.imshow(postprocessing_stencil, cmap='bwr', alpha=.2)
plt.show()
```
