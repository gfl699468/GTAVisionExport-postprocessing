```python
import glob
import os
from configparser import ConfigParser
import progressbar
from joblib import Parallel, delayed
from PIL import Image
import numpy as np
import csv
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
images_dir = '../depthEstimationNN1/data/make3d_datasets/Dataset2_Images'
images_depth_dir = '../depthEstimationNN1/data/make3d_datasets/depthMapData2Trans'
image_name = 'img-combined1-p-15t0.extraperson'
depth_name = 'depth-combined1-p-15t0.mat'
```
